## MP3 to Text Encoder

This program takes in mp3 files and converts them to text format to be used by the models.




The following Code is to encrypt one singular mp3 file to text

```python

import base64


#File path of input
filePathIn = "MusicMP3V2\Piano Sonata No. 1 In F Minor, Op. 2, No. 1_ 1. Allegro.mp3"

#File Path of Output
filePathOut = "MusicStringV2\Piano Sonata No. 1 In F Minor, Op. 2, No. 1_ 1. Allegro.txt"

#Reading in the data
with open(filePathIn, 'rb') as file:
    binaryData = file.read()
    
#To base64 then to text
encodedData = base64.b64encode(binaryData)
stringData = encodedData.decode('utf-8')

#Writing to output
with open(filePathOut, 'w') as file:
    file.write(stringData)

#print(stringData)
```

### This code is capable of converting an entire folder full of mp3 files to one singular text file

It is important to delete the previous version of AllSongs or to change the value stored in filePathOut prior to running, otherwise all of the text will be appended to the end of AllSongs.

```python
import os
import base64

#Directory housing mp3 files
directoryPath = "MusicMP3"
files = os.listdir(directoryPath)

#File path of output file
filePathOut = "MusicString\AllSongs.txt"

counter = 0
for fileName in files:
    filePath = os.path.join(directoryPath, fileName)
    
    if os.path.isfile(filePath):
        try:
            #Reading in the data
            with open(filePath, 'rb') as file:
                binaryData = file.read()

            #To base64 then to text
            encodedData = base64.b64encode(binaryData)
            stringData = encodedData.decode('utf-8')
            
            #Writing to output
            with open(filePathOut, 'a') as file:
                file.write(stringData)
                
            print(f"Appended {fileName} to {filePathOut}")
            counter += 1
            
        except IOError as e:
            print(f"Failed to append {fileName}: {e.strerror}")
            
print(f"Number of songs appended: {counter}")
```

### Decoder for one text file back to mp3

```python
import base64


#Text File
#filePathIn = 'MusicString\Piano Sonata No. 1 In F Minor, Op. 2, No. 1_ 1. Allegro.txt'
filePathIn = 'MusicString\AllSongs.txt'
#MP3 File
filePathOut = 'AllSongs1.mp3'

#Read Text
with open(filePathIn, 'r') as file:
    encodedData = file.read()

#Decode base64 string    
mp3_data = base64.b64decode(encodedData)

#Writing the base64 to mp3
with open(filePathOut, 'wb') as mp3_file:
    mp3_file.write(mp3_data)














```
